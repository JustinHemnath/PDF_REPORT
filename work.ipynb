```python
import json
import pandas as pd
from typing import List, Dict
from itertools import groupby
import math

with open("ward_alangulam.json", "r", encoding="utf-8") as f:
    res = json.load(f)
    idx_data = res["index_data"]

df = pd.DataFrame(idx_data)


class COMPARE_TYPE:
    AC = "ac"
    LOCAL_BODY = "local_body"
    BOOTH = "booth"
    LOCALITY = "locality"
    WARD = "ward_vp"

# df["ward_wise_booth_count"] = 0
# df["lb_wise_booth_count"] = 0
# df["trend_wise_booth_count"] = 0
df
```

```python
lb_df = df[df["lb_name"] == "முக்கூடல் பேரூர்"]
new_json = lb_df.to_dict(orient="records")

with open("new_json.json", "w", encoding="utf-8") as f:
    f.write(json.dumps(new_json, ensure_ascii=False))
```

```python
ward_list = df[["lb_name", "ward_vp"]]
new_ward_list = set(ward_list.to_records(index=None).tolist())
   
for lb_name, ward_vp in new_ward_list:
    matching_records_df = df[(df["lb_name"] == lb_name) & (df["ward_vp"] == ward_vp)]
    total = matching_records_df["booth_count"].sum()
    ward_item_list = matching_records_df.index.tolist()

    for idx in ward_item_list:
        idx_data[idx]["ward_wise_booth_count"] = total

# pd.DataFrame(idx_data)
pd.DataFrame(idx_data)
```

```python
## trend wise total booth count
trend_list_mapping = df[["lb_name", "ward_vp", "trend"]]
new_trend_list_mapping = set(trend_list_mapping.to_records(index=None).tolist())

for lb_name, ward_vp, trend in new_trend_list_mapping:
    matching_records_df = df[(df["lb_name"] == lb_name) & (df["ward_vp"] == ward_vp) & (df["trend"] == trend)]
    total = matching_records_df["booth_count"].sum()
    ward_item_list = matching_records_df.index.tolist()

    for idx in ward_item_list:
        idx_data[idx]["trend_wise_booth_count"] = total

pd.DataFrame(idx_data)
```
